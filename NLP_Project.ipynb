```python
from datasets import load_dataset
import os
```

```python
directory_path = "/content/drive/MyDrive/IN-Abs"
```

```python
train_judgements_path = "/content/drive/MyDrive/IN-Abs/train-data/judgement"
train_summary_path = "/content/drive/MyDrive/IN-Abs/train-data/summary"
```

```python
test_judgements_path = "/content/drive/MyDrive/IN-Abs/test-data/judgement"
test_summary_path = "/content/drive/MyDrive/IN-Abs/test-data/summary"
```

```python
def load_text_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        return f.read()
```

```python
train_judgement_files = [os.path.join(directory_path, 'train-data/judgement', file_name) for file_name in os.listdir(os.path.join(directory_path, 'train-data/judgement'))]
train_summary_files = [os.path.join(directory_path, 'train-data/summary', file_name) for file_name in os.listdir(os.path.join(directory_path, 'train-data/summary'))]
test_judgement_files = [os.path.join(directory_path, 'test-data/judgement', file_name) for file_name in os.listdir(os.path.join(directory_path, 'test-data/judgement'))]
test_summary_files = [os.path.join(directory_path, 'test-data/summary', file_name) for file_name in os.listdir(os.path.join(directory_path, 'test-data/summary'))]
```

```python
def preprocessing_function(example):
    preprocessed_example = {}
    for key, value in example.items():
        preprocessed_example[key] = load_text_file(value)
    return preprocessed_example
```

```python
train_dataset = load_dataset('text', data_files={'judgement': train_judgement_files, 'summary': train_summary_files}, split='train')
train_dataset = train_dataset.map(preprocessing_function)

test_dataset = load_dataset('text', data_files={'judgement': test_judgement_files, 'summary': test_summary_files}, split='test')
test_dataset = test_dataset.map(preprocessing_function)
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')

train_dataset = train_dataset.map(lambda examples: tokenizer(examples['judgement'], examples['summary']), batched=True)
test_dataset = test_dataset.map(lambda examples: tokenizer(examples['judgement'], examples['summary']), batched=True)
```
